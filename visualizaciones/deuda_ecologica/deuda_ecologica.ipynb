```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('Date.csv', sep=' ')
data
```

```python
french_months =  {'janvier':1,
                  'février':2,
                  'mars':3,
                  'avril':4,
                  'mai':5,
                  'juin':6,
                  'juillet':7,
                  'août':8,
                  'septembre':9,
                  'octobre':10,
                  'novembre':11,
                  'décembre':12}

data = data.replace({"month": french_months})
data
```

```python
data['year'] = data['year'].str.replace('*','', regex=False)
data
```

```python
data['date'] = pd.to_datetime(data['year'].astype(str)+'/'+data['month'].astype(str)+'/'+data['day'].astype(str))
data
```

```python
data['no_days'] = data['date'] - pd.to_datetime(data['year'].astype(str)+'/01/01')
data
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,12.)) # 531 × 418

# don't show lines delimiting the graph
for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)

# plot data
ax.plot(data['year'], data['no_days'].dt.days, color='#53B8BB', linewidth=3.4, alpha=1)


# change the color of the axes' tick labels
ax.tick_params(axis='x', colors='grey')
ax.tick_params(axis='y', colors='grey')

x_ticks_custom = np.arange(1970, 2021,5)
ax.set_xticks(x_ticks_custom-1970)
ax.set_xticklabels(x_ticks_custom, size=14)

y_ticks_custom = [200, 240, 280, 320, 360]
ax.set_yticks(y_ticks_custom)
y_tick_labels = ['JUL 19', 'AGO 28', 'OCT 07', 'NOV 16', 'DIC 26']
ax.set_yticklabels(y_tick_labels, size=13)


# title and description
ax.text(-8.5, 410-5, s='El Día del Sobregiro Ecológico ocurrió temprano \neste año', size=40.9, weight='bold')
ax.text(-8.5, 380-5, 
        s='El Día del Sobregiro Ecológico marca el día del año en el que consumimos la cantidad de recursos que el planeta \nes capaz de generar durante el mismo año. Debido a nuestro creciente consumo insostenible, este día ha ocurrido \ncada vez más temprano durante las últimas 5 décadas. Fuente: www.overshootday.org', 
        size=19.3, linespacing=1.3)

# authorship
ax.text(-10, 178, ' ' * 300,
                    color='white',
                    backgroundcolor='grey',
                    size=17)
ax.text(43.5, 178,'Twitter: @DiegoGonH',
                color='white',
                size=16)


ax.set_ylabel(r'Fecha del Sobregiro Ecológico', size=20, labelpad=27, color='grey')
plt.subplots_adjust(right=0.999999999, left=0.11, bottom=0.083, top=0.74)
plt.savefig('deuda_ecologica.jpg', dpi=128)

plt.show()
```
